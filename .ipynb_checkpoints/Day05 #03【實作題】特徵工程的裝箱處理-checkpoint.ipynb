```python
'''
#練習：請根據提供的 `iris` 資料，完成以下操作：

1. 利用資料視覺化觀察 sepal 和 petal 四個欄位的關係。
2. 利用相關係數觀察 sepal 和 petal 四個欄位的關係。
'''

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
ax1 = df[['sepal length (cm)','petal length (cm)','sepal width (cm)','petal width (cm)']]
plt.plot(ax1)
plt.legend(ax1)
ax1.corr()
```

```python
df[['sepal length (cm)','petal length (cm)','sepal width (cm)','petal width (cm)']].plot()
df.corr()
```
