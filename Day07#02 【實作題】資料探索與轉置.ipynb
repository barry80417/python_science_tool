```python
import numpy as np
import pandas as pd

ipl_data = {
    'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
    'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
    'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
    'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
    'Points':[876,789,863,673,741,812,756,788,694,701,804,690]
}
df = pd.DataFrame(ipl_data)

df
```

```python
# df['Team']=df['Team'].str.capitalize()  ＃首字大寫
df.groupby(["Team","Year"]).size()
```

```python
df.groupby(["Team","Year"]).size().unstack(fill_value = None) #unstack function是將橫向column轉成縱向
```

```python
pd.crosstab(df['Team'],df['Year']).replace(0,np.nan)
```

```python
df.pivot_table(index=['Team'], columns='Year', aggfunc = 'size')
```
